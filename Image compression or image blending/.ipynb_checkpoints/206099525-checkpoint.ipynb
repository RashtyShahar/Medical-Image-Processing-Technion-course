2. Implement the function `bulid_tree` that receives a dictionary of the symbols and their probabilities and returns a nested list of the image values representing the huffman tree. For example, given the dictionary: {0: 0.2, 1: 0.1, 2: 0.1, 3: 0.4, 4: 0.2} the tree should be [[0, 4], [[1, 2], 3]]. Write a description of your function and explain its inputs and output.

-*note*- </br>
For this assignment I used 3 functions:</br>
the first builds the tree in a dictionary form. (had to use tuples here because lists are unhashable in dictionaries).</br>
the second turns nested tuples into nested lists.</br>
the last one uses the both first ones to make a binary tree in the form of nested lists.

```python
def build_tree_tuples(dict_prob: dict) -> dict:
    '''
    this is a recursive function.
    it takes a dictionary with the length n+2, that has nomeric values.
    it returns a shorter dictionary with the length n+1.
    in each recursion it deletes the two key with the smallest values
    and adds a new key in a form of a tuple containing both delteted keys and their added values as value.
    the recursion stops when the dictionary has only two keys.

    :param dict_prob: a dictionary representing color intensities as keys and their probability as values
    :return tree: two keys long dictionary representing color intensities as keys and their probability as values
    '''
    if len(list(dict_prob.values())) == 2 :
        product = dict_prob
        return product
    else:
        list_prob = sorted(dict_prob.items(), key=lambda x:x[1] )
        sortdict = dict(list_prob[2:])
        sortdict.update({tuple([list_prob[0][0] , list_prob[1][0]]) : list_prob[1][1] + list_prob[2][1]})
        sortdict = build_tree_tuples(sortdict)
        return sortdict


def tuple_to_list(nested_tuple: tuple) -> list:
    '''
    recursive function that turns a nested tuple into a nested list in the same order.

    :param nested_tuple: nested tuple
    :return nested_tuple: nested list
    '''

    if type(nested_tuple) == tuple:
        return [tuple_to_list(i) for i in nested_tuple]
    return nested_tuple

def build_tree(dict_prob: dict) -> list:
    '''
    The function takes a dictionary representing color intensities as keys and their probability as value.
    It returns a binary tree in a form of a nested list.
    the tree represents the probability hierarchy between the different color intensities.
    
    :param dict_prob: a dictionary representing color intensities as keys and their probability as values
    :return tree:  binary tree in the form of a nested list.
    '''
    # ====== YOUR CODE: ======
    tree = tuple(build_tree_tuples(dict_prob))
    tree = tuple_to_list(tree)

    # ========================  
    return tree
```

3. Implement the function `huffman code` that receives the huffman tree and returns a dictionary of the symbols and their huffman coding (a binary string). <br> If needed, you may add an additional input to the function. 

    Optional guide: Assign an empty string to the root and then recursively assign 1 for the right branch and  0 for the left branch.<br>
    
    Write a description of your function and explain its inputs and output.

```python
def huffman_code(root: list , code: str ) -> dict:
    '''
    This is a recursive function.
    It gets a binary tree in the form of a nested list.
    It returns a dictionary containing the leafs as keys and the binary address for each leaf in the tree as values (as string type).
    :param huffman_tree: binary tree in the form of a nested list.
    :return dict_code: dictionary containing each leaf and it's binary 'address'
    '''
    # ====== YOUR CODE: ======
    if type(root) != list:
        return {root : code}
    else:
        dict_code = huffman_code(root[0] , code +'1')
        dict_code.update(huffman_code(root[1] , code + '0'))
    # ========================         
        return dict_code

```

4. Implement the function `huffman_encode` that receives the huffman code and the flatten image and returns the encoded image.
 Write a description of your function and explain its inputs and output.

```python
#ive changed the output into a string to match what was learned in class.
def huffman_encode(code: dict, img: np.ndarray) -> str:
    '''
    This function receives the huffman code  in a form of a dictionary and a flatten image.
    It returns the encoded image as a list of binary strings.

    :param code: dictionary containing the huffman code .
    :params img: a vector representing flatten image.
    :return encoded_img:  an str representing the encoded flatten image.
    '''
    # ====== YOUR CODE: ======
    encoded_img = ''
    for item in img:
        temp_word = code.get(item)
        encoded_img += (temp_word)

    # ========================
    return encoded_img
```

```python
#ive changed the input to be a string to match what was learned in class.

def huffman_decode(encoded_img:str, tree: list) -> np.array:
    '''
    The function receives the encoded image and the huffman tree of the image and returns the decoded image.
    
    :param encoded_img:  a string representing the flatten encoded image.
    :params tree: nested lists representing the binary tree that was used to make the huffman code.
    :return decoded_img: a vector representing flatten decoded image.

    '''
    # ====== YOUR CODE: ======
    decoded_img = []
    encode = huffman_code(tree , "")
    reverse_code = {}
    for key, value in encode.items():
        reverse_code.update({value : key})

    word = ''
    for bit in encoded_img:
        word += bit
        if reverse_code.get(word) != None:
            decoded_img.append(reverse_code.get(word))
            word = ''

    # ========================
    return decoded_img
```

In this exercise, you will implement and use an algorithm base on Laplacian pyramids used to stitch multiple images together. 

1. First, implement the `gaussian_pyramid`  function that creates a list of downscaled versions of the input image by a scale of 2. Before each down sampling, use gaussian smoothing to avoid aliasing.

    Note: The smoothing is required to avoid aliasing. Do not smooth the mask so that it will remain a binary mask.

    Do not use the resize functions for downscaling. You may use the 'gaussian' function from skimage.filters
    
    Write a description of your function and explain its inputs and output.


```python
def gaussian_pyramid(
        img: np.ndarray,
        num_levels: int,
        mask: bool = False,
        sigma: float = 0.4,
    ) -> list:
    '''
    The function recieves an image, number of levels and a sigma value and a bulian param.
    it then returns a list of images representing the gaussian pyramid.
    The list length is according to the number of levels input.
    each level goes through gaussian smoothing (according to sigma input) before being downscaled for the next level.
    if the mask input is False - their will be no gaussian smoothing preformed.

    :param img: (H,W,3) array representing an image
    :params num_levels: integer representing the number of levels wanted
    :params mask: bulian that clarifies the inputed image designation - mask or image
    :params sigma: std for gaussian smoothing
    :return gaussian_pyr: a list of arrays representing a gaussian pyramid
    '''
    # ====== YOUR CODE: ======
    temp_img = img
    gaussian_pyr = []
    if mask :
        for i in range(num_levels):
            gaussian_pyr.append(temp_img)
            temp_img = temp_img[::2,::2,:]
    else:
        for i in range(num_levels):
            gaussian_pyr.append(temp_img)
            temp_img = gaussian(temp_img ,sigma , channel_axis=True )[::2,::2,:]
    
    # ========================
    return gaussian_pyr
        
```

2. Implement the `laplacian_pyramid` function that receives a Gaussian pyramid and computes an approximation of the Laplacian using the difference of Gaussians in every two consecutive levels of the pyramid. The top of the pyramid (lowest resolution) will remain the same.

    You may use the 'resize' function from skimage.transform

    Write a description of your function and explain its inputs and output.
    
    **Bonus (5 points):**   
Instead of using the 'resize' function from skimage, use your image transformation functions from homework assignments 3. Only instead of rotation, perform scaling.


```python
def laplacian_pyramid(gaussian_pyr: list) -> list:
    '''
    The function takes a gaussian pyramid and returns a laplacian pyramid.
    
    :param gaussian_pyr:  a list of arrays representing a gaussian pyramid
    :return laplacian_pyr:  a list of arrays representing a laplacian pyramid
    '''
    # ====== YOUR CODE: ======
    laplacian_pyr = []
    for i in range(len(gaussian_pyr)-1):
        temp_img = gaussian_pyr[i] - resize(gaussian_pyr[i+1], gaussian_pyr[i].shape , anti_aliasing=False )

        laplacian_pyr.append(temp_img)
    laplacian_pyr.append(gaussian_pyr[-1])
    
    # ======================== 
    return laplacian_pyr
```

3. Implement the `blend` function that receives 3 Laplacian pyramids, those of the two images you want to stitch together and that of the mask, and blend the two pyramids according to the mask. At each level of resolution, the combined pyramid will contain the pixels in the mask (ones) from the first pyramid and the pixels from out of the mask (zeros) from the second pyramid.

    Write a description of your function and explain its inputs and output.


```python
def blend(laplacian_A: list, laplacian_B: list, mask_pyr: list) -> list:
    '''
    This function recieves 2 laplacian pyramids of images and a third laplacian pyramid of a mask.
    It returns a pyramid with the stiched images of both pyramids down according to the mask.
    
    :param laplacian_A: a list of arrays representing a laplacian pyramid of image 1
    :params laplacian_B: a list of arrays representing a laplacian pyramid of image 2
    :params mask_pyr: a list of arrays representing a laplacian pyramid of the mask
    :return blended_imgs: a list of arrays representing a laplacian pyramid of the blended image
    '''
    # ====== YOUR CODE: ======
    blended_imgs = []
    for i in range(len(mask_pyr)):
        blended_imgs.append(laplacian_A[i] * mask_pyr[i] + laplacian_B[i] * (1- mask_pyr[i]))

    # ========================  
    return blended_imgs
```

4. Implement the `reconstruct` function that receives a Laplacian pyramid and reconstructs the Gaussian pyramid that was used to construct it by upscaling the top of the pyramid (lowest resolution) and adding it to the next layer and doing the same with the first reconstructed level to reconstruct the next layer and so on for all layers.

     You may use the 'resize' function from skimage.transform
     
     Write a description of your function and explain its inputs and output.


```python
def reconstruct(laplacian_pyr: list,) -> list:
    '''
    This function reconstructs an image out of its laplacian pyramid.
    
    :param laplacian_pyr: a list of arrays representing a laplacian pyramid
    :return recon_pyr: a (H,W,3) array representing the reconstructed image
    '''
    # ====== YOUR CODE: ======
    recon_pyr = laplacian_pyr[-1]
    for i in range(len(laplacian_pyr)-2,0,-1):
        recon_pyr = laplacian_pyr[i-1] + resize(recon_pyr, laplacian_pyr[i-1].shape , anti_aliasing=True )
    # ========================
    return recon_pyr        
```
