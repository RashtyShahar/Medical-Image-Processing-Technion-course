In many cases, you might want to blur your image before using algorithms to process it.  
A good kernel you can use to blur an image is the Gaussian kernel.  
Implement a function `gaussian_kernel` that returns a Gaussian kernel for a given kernel size and std.
Write a description of your function and explain its inputs and output.

```python
def gaussian_kernel(size: int, sigma: float) -> np.ndarray:
    '''
    returns a Gaussian kernel for a given kernel size and std 
    
    :param size: kernel size
    :param sigma: standad deviation of kernel
    :return gauss_kernel: Gaussian kernel for a given kernel size and std
    '''
    # ====== YOUR CODE: ======
    ax = np.linspace(-(size - 1) / 2., (size - 1) / 2., size)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-0.5 * (np.square(xx) + np.square(yy)) / np.square(sigma))
    gauss_kernel = kernel / np.sum(kernel)
    # ========================
    return gauss_kernel
```

   Implement the function `img_grad_sobel` that calculates the image gradient magnitude and direction. This function should return two arrays; one of the gradient magnitude and another of the gradient direction in radians. The gradient maps should have the same size as the input.  <br>
    Note: to avoid producing edges at the image's borders, use the reflection boundary condition.<br>
    You may use the `convolve2d` function of the scipy.signal package.  
    Write a description of your function and explain its inputs and output. 

 

```python
def img_grad_sobel(img: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    '''
    calculates the image gradient magnitude and direction
    :param img: input image as numpy array
    :return magnitude: array of the gradient magnitude
    :return direction:array of the gradient direction in radians
    '''
    
    # ====== YOUR CODE: ======
    sobel_x=np.asarray([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
    sobel_y=np.asarray([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
    d_x = signal.convolve2d(img, sobel_x,mode='same',boundary='symm')
    d_y = signal.convolve2d(img, sobel_y,mode='same',boundary='symm')
    magnitude = np.zeros_like(img)
    direction = np.zeros_like(img)
    w,h = img.shape[0:2]
    for i in range(w):
        for j in range(h):
            magnitude[i, j] = np.sqrt(d_x[i, j]**2+d_y[i, j]**2)
            direction[i, j] = np.arctan2(d_y[i, j], d_x[i, j])

    # ========================
    
    return magnitude, direction
```

3. The gradient magnitude of the blurred image contains thick blurred edges.   
We would like to make the edges in the gradient map thinner.   
Implement the function `non_max_suppression` that takes in a gradient image and extracts thin edges using non-maximum suppression.  
Your function has to go over all the pixels in the image and replace with 0 the pixels whose neighbors in the gradient direction have higher values. You can assume a neighborhood of 3x3. 
Since the edges of the image might not have a pixel to compare, you can change those pixels to 0.  
Write a description of your function and explain its inputs and output.

```python
def non_max_suppression(mag: np.ndarray, angle2: np.ndarray) -> np.ndarray:
    '''
    extract thin edges using non_maximum suppresion.
    At each  iteration we will check 1 of the 8 close neighboors according to gradient direction
    go over all the pixels in the image and replace with 0 the pixels whose neighbors in the gradient direction have higher values.
    
    :param grad_magnitude: mag : the gradient magnitude
    :param grad_direction: angle2 : the gradient direction
    :return non_max: magnitude img with thiner edges 
    
    '''
    # ====== YOUR CODE: ======
    w, h = mag.shape
    non_max=np.zeros_like(mag)
    angle = angle2 * 180. / np.pi #convert radians to degrees
    angle[angle<0]+=180
    for i in range(1,w-1):
        for j in range(1,h-1):
           #angle 0 --> walk right or left on the same row
            if (0 <= angle[i,j] < 22.5) or (157.5 <= angle[i,j] <= 180):
                p = mag[i, j+1]
                r = mag[i, j-1]
            #angle 45 --> walk on the digonal right up direction
            elif (22.5 <= angle[i,j] < 67.5):
                p = mag[i+1, j-1]
                r = mag[i-1, j+1]
            #angle 90--> walk up and down at the same column
            elif (67.5 <= angle[i,j] < 112.5):
                p = mag[i+1, j]
                r = mag[i-1, j]
            #angle 135 --> walk on the digonal left down direction
            elif (112.5 <= angle[i,j] < 157.5):
                p = mag[i-1, j-1]
                r = mag[i+1, j+1]

            if (mag[i,j] >= p) and (mag[i,j] >= r): #thats the edge!
                non_max[i,j] = mag[i,j]
            else: #there is a bigger value at the gradient dirction --> thats not an edge
                non_max[i,j] = 0
            

    
    # ========================
                
    return non_max
```

5. After thinning the edges, we're left with high and low magnitude edges. We would like to filter out all the insignificant edges.
Implement the function `double_threshold` that gets an edge image and two thresholds and returns an image containing three values: 0 for background, 1 for weak edges, and 2 for strong edges.<br>
Write a description of your function and explain its inputs and output.

```python
def double_threshold(edges: np.ndarray, lowThresholdRatio: float, highThresholdRatio: float) -> np.ndarray:
    '''
    gets an edge image and two thresholds and returns an image containing three values:
    0 for background, 1 for weak edges, and 2 for strong edges.
    
    :param edges: edge image
    :param lowThresholdRatio: lower threshold - values smaller then this will be replaced with 0
    :param highThresholdRatio: higher threshold - values larger then this will be replaced with 2
    :return img_tresh:
    '''
    # ====== YOUR CODE: ======
    img_thresh = np.zeros_like(edges)
    w , h = edges.shape
    for i in range(w):
        for j in range(h):
            if edges[i][j]<=lowThresholdRatio:
                img_thresh[i][j] = 0
            elif edges[i][j] >= highThresholdRatio :
                img_thresh[i][j] = 2
            else :
                img_thresh[i][j] = 1
            
    # ========================
    return img_thresh
```

```python
def hysteresis(edges: np.ndarray) -> np.ndarray:
    '''
    filtering the isolated weak edges and keeping only the strong edges and those that are connected to them.
    
    :param edges: edges image after applying double_threshold (values can be 0 / 1 / 2)
    :return hysteresis: image's real edges according to Canny edge detector
    '''
    # ====== YOUR CODE: ======
    hysteresis = edges.copy()
    w , h = edges.shape
    for i in range(1,w-1):
        for j in range(1,h-1):
            neighbors = [(i+1,j),(i-1,j),(i,j+1),(i,j-1),(i+1,j+1),(i+1,j-1),(i-1,j-1),(i-1,j+1)]
            if hysteresis[i][j]==1: # if true we need to check neighborhood
                for tup in neighbors:
                    if hysteresis[tup]==0 or hysteresis[tup]==1: #if true, neighbor is not an edge
                        hysteresis[i][j]=0
                        continue 
                    elif hysteresis[tup]==2: #if true, neighbor is an edge so current pixel will be set to edge as well 
                        hysteresis[i][j]=2
                        break  #if we found an edge neighbor no need to check the rest       
    # ========================    
    return hysteresis
```

2. For Hough Circle Transform we need three parameters to define a circle: $(x_{center}, y_{center}, r)$ <br>
where $(x_{center},y_{center})$ define the center position and $r$ is the radius.
It means that in this case the accumolator matrix would be a 3D matrix.<br>
Implenent the function `circles_accumolator` that gets a binary image that repersents the image edges and an array of the different radiuses to search circles at, and return the accomultor matrix in the hough space. 


```python
def circles_accumolator(img:np.ndarray, radius:np.ndarray)-> np.ndarray:
    '''
     gets a binary image that repersents the image edges and an array of the different radiuses to search circles at,
     and return the accomultor matrix in the hough space
     
    :param img: binary image that repersents the image edges
    :param radius: array of the different radiuses to search circles at
    :return hough_domain: the accomultor matrix in the hough space
    '''
    # ====== YOUR CODE: ======
    M,N = img.shape
    R_max = max(radius)
    hough_domain = np.zeros((M,N,R_max+1))
#     hough_domain = np.zeros((M,N,len(radius)))
    pixels_indexs = np.argwhere(img!=0)
    angles = np.arange(-180,180,15)  
#     radians = list(map(lambda x : float(x/180*np.pi) , angles))
    radians = (np.pi/180)*angles
    for xi,yi in pixels_indexs:
        for r in radius:
            for rad in radians:
                a = int(xi + r*np.cos(rad))
                b = int(yi + r*np.sin(rad))
                if 0<=a<M and 0<=b<N:
                    hough_domain[a, b, r] += 1    
    # ========================    
    return hough_domain 
```

4. Implement the function `hough_circles` that gets image of edges, computes the accomultor matrix in the hough space and return the parameters (a,b,r) of each detected circle on the image. Choose the detected circle acorrding to votes threshold and distance thresholds that represents the minimal distance between the centers of two different circles. 

```python
def hough_circles(img:np.ndarray, radius:np.ndarray, votes_thresh:int, dist_thresh:float)-> np.ndarray:
    '''
    gets image of edges, computes the accomultor matrix in the hough space and return the parameters (a,b,r) of each detected
    circle on the image that passes the votes and distance thresholds.
    
    :param img: image of edges
    :param radius: array of the different radiuses to search circles at
    :param votes_thresh: votes threshold - integer
    :param dist_thresh: distance thresholds represents the minimal distance between the centers of two different circles
    :return circles:  the parameters (a,b,r) of each detected circle on the image
    '''
    
    # ====== YOUR CODE: ======
    hough_domain  = circles_accumolator(img,radius)
    circles = []
    possible_circles_indexs=np.argwhere(hough_domain>votes_thresh)
    tempx,tempy,tempr=[],[],[]

    for i in range(len(possible_circles_indexs)-1):
        x,y,r = possible_circles_indexs[i][0],possible_circles_indexs[i][1],possible_circles_indexs[i][2]
        nextx , nexty ,nextr = possible_circles_indexs[i+1][0],possible_circles_indexs[i+1][1],possible_circles_indexs[i+1][2]
        if np.hypot((nextx-x),(nexty-y))<dist_thresh:
            tempx.append(x)
            tempy.append(y)
            tempr.append(r)
        elif tempx!=[]:
            circles.append((np.mean(tempx),np.mean(tempy),np.mean(tempr)))
            tempx,tempy,tempr=[],[],[]
#             circles.append((np.mean(possible_circles_indexs[i][0]) , np.mean(possible_circles_indexs[i][1]) ,np.mean(possible_circles_indexs[i][2])))                   
#             circles.append((x,y,r))
    # ========================   
    return circles
```
